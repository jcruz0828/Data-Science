```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic_df = pd.read_csv('Data Science/Data/titanic_train.csv') 
```

```python
titanic_df['Pclass'].value_counts().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.show() 
```

```python
titanic_df['Age'].hist(bins=20)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show() 
```

```python
titanic_df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate')
plt.ylabel('')
plt.show() 
```

```python
sns.boxplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Fare Distribution by Class')
plt.show() 
```

```python
sns.violinplot(x='Survived', y='Age', data=titanic_df)
plt.title('Age Distribution by Survival')
plt.show() 
```

```python
sns.heatmap(titanic_df.corr(), annot=True, cmap='coolwarm') 
plt.title('Correlation Matrix')
plt.show() 
```
